## Уровень 0
посчитайте значение 1-ой и 2-ой производной фунции
f(x) = x^5 + 4sin(2x) + cos(3x+3) в точке x = 1

```python
import numpy as np
import scipy.misc
import matplotlib.pyplot as plt
```

```python
def func_1(x):
    return x**5 + 4*np.sin(2*x) + np.cos(3*x+3)

x = np.linspace(-100, 100, 500)
y = np.array([func_1(x_val) for x_val in x])

plt.plot(x, y)
plt.title('График функции f(x) = x^5 + 4sin(2x) + cos(3x+3)', pad=30)
plt.grid()
plt.show()
```

```python
x = 1
print(f'Значение первой производной функции в точке x={x}:', scipy.misc.derivative(func_1, 1, n=1, dx=1e-6))
print(f'Значение второй производной функции в точке x={x}:', scipy.misc.derivative(func_1, 1, n=2, dx=1e-6))
```

## Уровень 1

Постройте график первой и второй производной функции
(sin(2x+1))^5
на отрезке [-5; 5]

```python
def func_2(x):
    return np.sin(2*x+1)**5

x = np.linspace(-5, 5, 100)
y = np.array([func_2(x_val) for x_val in x])

derivative_n1 = scipy.misc.derivative(func_2, x, n=1, dx=1e-6)
derivative_n2 = scipy.misc.derivative(func_2, x, n=2, dx=1e-6)

plt.plot(x, y, label='Функция')
plt.plot(x, derivative_n1, '--r', label='Первая производная')
plt.plot(x, derivative_n2, '--y', label='Вторая производная')
plt.title('График функции f(x) = (sin(2x+1))^5 и ее производных', pad=15)
plt.grid()
plt.legend()
plt.show()
```
